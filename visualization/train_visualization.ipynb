```python
import os
import plotly.graph_objs as go
from tensorboard.backend.event_processing import event_accumulator

monitor = 'val_minADE'
use_case = 'D4' #A1/B1/C1/D3/D4
step = [1, 5, 10]

fig = go.Figure()

for st in step:
    log_dir = os.path.join('./logs', '{}_step{}'.format(use_case, st))
    ea = event_accumulator.EventAccumulator(log_dir)
    ea.Reload()

    scalar_data = ea.Scalars(monitor)

    steps = [s.step for s in scalar_data]
    values = [s.value for s in scalar_data]

    fig.add_trace(go.Scatter(x=steps, y=values, mode='lines', name=f'{use_case} - sampling{st} - {monitor}'))

fig.update_layout(
    title="Validation Minimum Average Displacement Error Over Steps",
    xaxis_title="steps",
    yaxis_title="minADE",
    template='plotly_dark'  # 어두운 테마 설정
)

fig.show()
```
